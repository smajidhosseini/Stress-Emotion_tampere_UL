```python
# import libraries
import cv2
import time
import dlib
import json
import os
import sys
#import torch
import tqdm as tqdm
from bokeh.plotting import figure
from IPython.display import display, Image
#device = torch.device('cuda') if torch.cuda.is_available() else 'cpu'
```

```python
root = 'E:\\Users\\Python\\Desktop\\Emotion_Data'
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
for sess in (sorted(os.listdir(root))):   
    for filename in os.listdir(os.path.join(root, sess)):
           
        if filename.endswith('.mp4'):
            print(os.path.join(root, sess,filename))

            cap= cv2.VideoCapture(filename)
            pTime=0   

            index = 0
            d = {"frame": {'No': {"x": "", "y": ""}}}
            index = 0
            while True:

                # Capture frame-by-frame
                ret, frame = cap.read()
                if (ret == False):
                    sys.stdout.write("no image: %d%%   \r" % (index) )
                    sys.stdout.flush()
                    break
                cTime = time.time()
                fps= 1/ (cTime-pTime)
                pTime=cTime
                #cv2.putText(frame,f'FPS:{int(fps)}',(20,70),cv2.FONT_HERSHEY_COMPLEX,3,(0,255,0),3)
                #cv2.waitKey(1)
                if (index < 16200):
                    faces = detector(frame)
                    for face in faces:
                        f = {}
                        x1 = face.left()
                        y1 = face.top()
                        x2 = face.right()
                        y2 = face.bottom()
                        #print(x1,end="")
                        cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),3)
                        landmarks = predictor(frame,face)
                        for n in range(68):
                            x= landmarks.part(n).x
                            y= landmarks.part(n).y
                            f.update({n:{'x':x,'y':y}})
                        d.update({index:f})
                            # Display the resulting frame
                    cv2.imshow('Frame', frame)
                    cv2.waitKey(1)
                    # Press Q on keyboard to  exit
                    #if cv2.waitKey(25) & 0xFF == ord('q'):
                    #    print('hi')
                    #    break

                    # Break the loop

                    index+=1
                    sys.stdout.write("frmaes processed: %d%%   \r" % (index) )
                    sys.stdout.flush()
                else:
                    break
            # When everything done, release 
            # the video capture object
            cap.release()

            # Closes all the frames
            cv2.destroyAllWindows()
            
            with open(os.path.join(root, sess,filename)+'.json', "w") as outfile:
                print(f'{os.path.join(root, sess,filename)}'+'.json')
                json.dump(d, outfile)
```
